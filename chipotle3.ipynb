```python
# column 단위 데이터에 apply()함수로 전처리를 적용한다.
chipo['item_price'].apply(lambda x : float(x[1:]))
chipo.describe()
```

```python
# processin
chipo['order_id'] = chipo['order_id'].astype('str')
chipo['item_price'] = chipo['item_price'].apply(lambda x : float (x[1:]))
chipo.head()
```

```python
chipo.info()
```

```python
chipo.groupby('order_id')['item_price'].sum()
```

```python
# 주문당 평균 계산금액을 출력한다.
chipo.groupby('order_id')['item_price'].sum().mean()
```

```python
# 한 주문에 30달러 이상 지불한 주문 번호('order_id') 출력하기
chipo_orderid_group = chipo.groupby('order_id').sum()
results = chipo_orderid_group[chipo_orderid_group.item_price >= 30]
print(results[:10])
print(results.index.values)
```

```python
# 각 아이템의 가격 구하기
chipo_one_item = chipo[chipo.quantity == 1]
chipo_one_item.groupby('item_name').min()
```

```python
# sort_values() 함수를 적용하여 데이터를 정렬한다.
price_per_item = chipo_one_item.groupby('item_name').min()
price_per_item.sort_values(by = 'item_price',ascending = False)
```
